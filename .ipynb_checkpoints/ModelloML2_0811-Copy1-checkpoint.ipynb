```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=10,5

from scipy.integrate import solve_ivp
from ode_helpers import state_plotter
```

```python
#TEST FUNCTION: dy_0/dt = 4cos(3t) and dy_1/dt = -2y_0 + 0.5t

#Define derivative function
def f(t, y, c):

    dydt = [c[0]*np.cos(c[1]*t), c[2]*y[0]+c[3]*t]

    return dydt

#Define time spans, initial values, and constants
tspan = np.linspace(0, 5, 100)
yinit = [0, -3]
c = [4, 3, -2, 0.5]


#Solve differential equation
sol = solve_ivp(lambda t, y: f(t, y, c), 

                [tspan[0], tspan[-1]], yinit, t_eval=tspan, rtol = 1e-5)

#Plot states
state_plotter(sol.t, sol.y, 1)
```

```python
# VALIDATION OF THE CANCER CELL GROWTH DYNAMICS MODEL (ART 10)
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def odes(x, t):
    # constants
    c = [0.07, 4e6, 3.7e-8, 2e3, 0.231, 0.017]
    #     r     K    muA     a    muC     d
    #     0     1     2      3     4      5
    
    dose = [0, 6.25, 3.125, 1.5625, 0.78, 0.39, 0.195, 0.098, 0.049, 0.024, 0.012, 0.006]
    muAC = [0, 0.046, 0.0322, 0.0225, 0.0158, 0.011, 0.0077, 0.0054, 0.0038, 0.0026,
            0.0018, 0.0013] #non parla nella tabella. Hp 0
    muCA = [0, 0.46, 0.322, 0.225, 0.158, 0.11, 0.077, 0.054, 0.038, 0.026, 0.018, 0.013]

    # assign each ODE to a vector element
    A = x[0]
    D = x[1]
    C = x[2]

    # define each ODE
    # alive
    dAdt = c[0]*x[0]*(1-x[0] / c[1]) - c[2]*x[0]*x[1] - ((muAC[0]*x[0]*x[2]) / (c[3] + x[2]))
    # dead
    dDdt = c[2]*x[0]*x[1] - c[5]*x[1] + ((muAC[0]*x[0]*x[2]) / (c[3] + x[2]))
    # drug
    dCdt = -c[4]*x[2] - ((muCA[0]*x[2]*x[0]) / (c[3] + x[2]))

    return [dAdt, dDdt, dCdt]

# initial conditions
x0 = [5e4, 2500, dose[0]*6e17]

# declare a time vector (time window)
t = np.linspace(48, 312, 23)
x = odeint(odes,x0,t)

A = x[:,0]
D = x[:,1]
C = x[:,2]

# experimental values
t_exp = [48, 60, 72, 84, 96, 108, 120, 132, 144, 156, 168, 180, 
         192, 204, 216, 228, 240, 252, 264, 276, 288, 300, 312]
val_exp = [5e4, 7e4, 1e5, 1.2e5, 5.9e5, 1.85e6, 2.58e6, 3.49e6, 
           3.65e6, 2.44e6, 2.32e6, 2.02e6, 1.71e6, 1.55e6, 1.45e6, 
           1.24e6, 9.6e5, 8.7e5, 8.5e5, 7.2e5, 5e5, 2.1e5, 5e4]
val_guzev = [5.4e4, 122897, 273025, 577722, 1114930, 1862730, 
             2621770, 3180420, 3498490, 3641320, 3654730, 3500150, 
             1062170, 2364930, 1672210, 1149160, 779597, 551455, 388608, 
             276498, 198206, 142884, 103435]

# plot the results
plt.plot(t,A, label = "Simulated alive")
plt.plot(t,D, label = "Simulated dead")
plt.plot(t_exp, val_exp, 'ko', label = "Experimental alive")
plt.xlabel("time (h)")
plt.ylabel("no. cells/mL")
plt.legend()
plt.title("Model Validation without drugs")

plt.show()
```

```python
# error = ( |experimental - ideal| / ideal ) * 100

error = []
error_g = []

# considering the error between simulated data with experimental 
#values and guzev's simulation data
for i in range(len(t)):
    #EXPERIMENTAL ERROR
    error.append((np.absolute(val_exp[i] - A[i]) / A[i]) * 100)
    #GUZEV ERROR
    error_g.append((np.absolute(val_guzev[i] - A[i]) / A[i]) * 100)

plt.plot(t, error, label = "Experimental")
plt.plot(t, error_g, label = "Guzev")
plt.xlabel("time (h)")
plt.ylabel("error [%]")
plt.legend()

print (f"Experimental average error: {round(np.average(error), 3)} %")
print (f"Guzev average error: {round(np.average(error_g), 3)} %")
```

```python
# VALIDATION OF THE CANCER CELL DRUG CYTOTOXICITY DYNAMICS MODEL (ART 10)
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def f(t, x, i):
    # assign each ODE to a vector element
    #alive
    A = c[0]*x[0]*(1 - (x[0] / c[1])) - c[2]*x[0]*x[1] - ((muAC[i]*x[0]*x[2]) / (c[3] + x[2]))
    #dead
    D = c[2]*x[0]*x[1] - c[5]*x[1] + ((muAC[i]*x[0]*x[2]) / (c[3] + x[2]))
    #drug
    C = -c[4]*x[2] - ((muCA[i]*x[2]*x[0]) / (c[3] + x[2]))
    
    dxdt = [A, D, C]
    return dxdt

# constants
c = [0.07, 4e6, 3.7e-8, 2e3, 0.231, 0.017]
#     r     K    muA     a    muC     d
#     0     1     2      3     4      5

muAC = [0, 0.046, 0.0322, 0.0225, 0.0158, 0.011, 0.0077, 0.0054, 0.0038, 0.0026,
            0.0018, 0.0013] #non parla nella tabella. Hp 0
muCA = [0, 0.46, 0.322, 0.225, 0.158, 0.11, 0.077, 0.054, 0.038, 0.026, 0.018, 0.013]
dose = [0, 6.25, 3.125, 1.5625, 0.78, 0.39, 0.195, 0.098, 0.049, 0.024, 0.012, 0.006]
# declare a time vector (time window)
t = np.linspace(0, 72, 72)
    
#error
error = []
sol_guzev = [1646260, 85683, 168014, 291671, 452951, 637807, 827114, 1003650,
            1156420, 1281250, 1378920, 1452970]
for i in range(len(muAC)):
    # initial conditions
    x0 = [2e4, 2500, dose[i]*6e17]

    sol = solve_ivp(lambda t, x: f(t, x, i), 
                    [t[0], t[-1]], x0, method = "RK45", t_eval = t)
    #print (sol.y[0])
    error.append((np.absolute(sol.y[0][71] - sol_guzev[i]) / sol_guzev[i]) * 100) 
    
    # plot the results
    plt.plot(sol.t, sol.y[0], label = dose[i])
    plt.xlabel("time (h)")
    plt.ylabel("no. cells/mL")
    
plt.title ("Alive with with drugs over time")
plt.legend()    
plt.show()
```

```python
### CELLULE VIVE INIZIALI +10% E +100%

## CAMBIO RISOLUTORE (su lungo termine)

from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def f(t, x):
    # assign each ODE to a vector element
    #alive
    A = c[0]*x[0]*(1 - (x[0] / c[1])) - c[2]*x[0]*x[1] 
    #dead
    D = c[2]*x[0]*x[1] - c[5]*x[1] 
    
    dxdt = [A, D]
    return dxdt

# constants
c = [0.07, 4e6, 3.7e-8, 2e3, 0.231, 0.017]
#     r     K    muA     a    muC     d
#     0     1     2      3     4      5

muAC = [0, 0.046, 0.0322, 0.0225, 0.0158, 0.011, 0.0077, 0.0054, 0.0038, 0.0026, 0.0018, 0.0013] 
muCA = [0, 0.46, 0.322, 0.225, 0.158, 0.11, 0.077, 0.054, 0.038, 0.026, 0.018, 0.013]

# declare a time vector (time window)
t = np.linspace(48, 312, 1000000)

# experimental values
t_exp = [48, 60, 72, 84, 96, 108, 120, 132, 144, 156, 168, 180, 
         192, 204, 216, 228, 240, 252, 264, 276, 288, 300, 312]
val_exp = [[5e4, 7e4, 1e5, 1.2e5, 5.9e5, 1.85e6, 2.58e6, 3.49e6, 
           3.65e6, 2.44e6, 2.32e6, 2.02e6, 1.71e6, 1.55e6, 1.45e6, 
           1.24e6, 9.6e5, 8.7e5, 8.5e5, 7.2e5, 5e5, 2.1e5, 5e4],
          [1e4, 1e4, 1e4, 3e4, 8e4, 8e4, 32e4, 80e4, 119e4, 312e4, 307e4, 295e4, 276e4, 212e4, 202e4, 190e4, 154e4,
          124e4, 103e4, 98e4, 75e4, 53e4, 16e4],
          [1e4, 1e4, 1e4, 2e4, 3e4, 4e4, 12e4, 13e4, 55e4, 82e4, 103e4, 130e4, 144e4, 160e4, 177e4, 198e4, 216e4,
          205e4, 190e4, 144e4, 103e4, 88e4, 24e4]]
val_guzev = [[5.4e4, 122897, 273025, 577722, 1114930, 1862730, 2621770, 3180420, 3498490, 3641320, 3654730, 3500150, 
             1062170, 2364930, 1672210, 1149160, 779597, 551455, 388608, 276498, 198206, 142884, 103435],
            [10070, 22602, 50236, 109941, 234397, 477462, 900296, 1509560, 2184030, 2732350, 3052090,
            3162600, 3125520, 2990300, 2785060, 2523550, 2212900, 1859650, 1474810, 1078450, 702772, 388800, 171158],
            [5228, 8821, 14854, 24948, 41736, 69424, 114512, 186545, 298370, 464737, 697464, 996212, 1338190, 1676040,
            1951030, 2114600, 2141610, 2029460, 1791020, 1451290, 1050550, 648799, 318579]]

A0 = [10e4, 2e4, 2e3]
color = ['bo--', 'ro--', 'go--']
error = [[],[],[]]
error_g = [[],[],[]]
for i in range(3):
    # initial conditions
    x0 = [A0[i], 2500]

    sol = solve_ivp(lambda t, x: f(t, x), 
                    [t[0], t[-1]], x0, method = "RK45", t_eval = t,
                    rtol = 2.220446049250313e-14, max_step = 0.9)
    
    for j in range (len(t_exp)):
        #EXPERIMENTAL ERROR
        error[i].append((np.absolute(val_exp[i][j] - sol.y[0][j*43478]) / sol.y[0][j*43478]) * 100)
        #GUZEV ERROR
        error_g[i].append((np.absolute(val_guzev[i][j] - sol.y[0][j*43478]) / sol.y[0][j*43478]) * 100)

    # plot the results
    plt.plot(sol.t, sol.y[0], label = '{:.1E}'.format(A0[i]))
    plt.plot(t_exp, val_exp[i], color[i], label = 'Experimental {:.1E}'.format(A0[i]))
    plt.xlabel("time (h)")
    plt.ylabel("no. cells/mL")
    
plt.title("Model Validation without drugs")
plt.legend()    
plt.grid()

plt.show()
```

### Risultato modifica cellule vive iniziali:
- 10%: non visibile
- 100%: spostamento a sx del grafico
